# Mouse Movement Plot for All Rows of a User from CSV
Bu notebook, seçilen profile_guid için CSV dosyasındaki tüm dakikalık mouse hareketlerini çizer.

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
# CSV yolunu belirt (dosya yolunu kendi yapına göre değiştir)
csv_path = "../data/activity_data.csv"
df = pd.read_csv(csv_path)
```

```python
# Kullanıcıya göre filtreleme
profile_guid = "002e8f20-d5c2-4153-b1d0-9e1332549ac6"
df_user = df[df['profile_guid'] == profile_guid]
```

```python
# Her satır için mouse hareketini çiz
for idx, row in df_user.iterrows():
    try:
        x_coords = ast.literal_eval(row['x'])
        y_coords = ast.literal_eval(row['y'])

        plt.figure(figsize=(8, 5))
        plt.plot(x_coords, y_coords, marker='o', linestyle='-')
        plt.title(f"Mouse Movement\n{profile_guid}\nStart: {row['start_date_time']}")
        plt.xlabel("X Coordinate")
        plt.ylabel("Y Coordinate")
        plt.gca().invert_yaxis()
        plt.grid(True)
        plt.show()
    except Exception as e:
        print(f"Hata oluştu satır {idx}: {e}")
```
