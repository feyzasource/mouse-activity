# Fake Veri - Mouse Movement Plot (İlk 480 Satır)
Bu notebook, `fake_activity_log.csv` dosyasındaki ilk 480 satırın mouse hareketlerini çizer.
Profil GUID dikkate alınmaz.

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
# 📄 CSV dosyasını oku
csv_path = "../../data/fake/fake_activity_log.csv"  # Jupyter'dan göreli yol
df = pd.read_csv(csv_path)
```

```python
# 🔁 İlk 480 satırın mouse hareketini çiz
for idx, row in df.head(480).iterrows():
    try:
        if pd.isna(row['x']) or pd.isna(row['y']):
            raise ValueError("x veya y boş")
        
        x_str = row['x']
        y_str = row['y']

        if '{' in x_str and not x_str.startswith('['):
            x_str = '[' + x_str.strip('{}') + ']'
        if '{' in y_str and not y_str.startswith('['):
            y_str = '[' + y_str.strip('{}') + ']'

        x_coords = ast.literal_eval(x_str)
        y_coords = ast.literal_eval(y_str)

        if not isinstance(x_coords, list) or not isinstance(y_coords, list):
            raise ValueError("x veya y list değil")
        if len(x_coords) != len(y_coords):
            raise ValueError("x ve y uzunlukları eşit değil")

        plt.figure(figsize=(8, 5))
        plt.plot(x_coords, y_coords, marker='o', linestyle='-')
        plt.title(f"Fake Mouse Movement (Satır {idx})")
        plt.xlabel("X Coordinate")
        plt.ylabel("Y Coordinate")
        plt.gca().invert_yaxis()
        plt.grid(True)
        plt.show()
        
    except Exception as e:
        print(f"Satır {idx} atlandı: {e}")
```
