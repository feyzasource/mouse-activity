```python
import pandas as pd
import ast
import numpy as np

# Step 1: Load the CSV file
file_path = r'D:\store\dev\berqun\ai\data\acer\processed\km_stat_acer_20250414_p_1.csv'
df = pd.read_csv(file_path)
# === Step 2: Defensive parsing of both x and y columns ===
def parse_array_pair(row):
    try:
        x = ast.literal_eval(row['x']) if isinstance(row['x'], str) else []
        y = ast.literal_eval(row['y']) if isinstance(row['y'], str) else []
        if isinstance(x, list) and isinstance(y, list) and len(x) == len(y):
            return pd.Series({'x': x, 'y': y})
        else:
            return pd.Series({'x': np.nan, 'y': np.nan})  # Invalidate both if mismatch
    except Exception:
        return pd.Series({'x': np.nan, 'y': np.nan})  # Invalidate if parsing fails

# Apply to each row
df[['x', 'y']] = df.apply(parse_array_pair, axis=1)
```

```python
def calc_total_distance(x, y):
    if isinstance(x, list) and isinstance(y, list) and len(x) >= 2:
        dx = np.diff(x)
        dy = np.diff(y)
        return np.sum(np.sqrt(dx**2 + dy**2))
    return np.nan

def calc_path_linearity(x, y):
    if isinstance(x, list) and isinstance(y, list) and len(x) >= 2:
        total_distance = calc_total_distance(x, y)
        euclidean_dist = np.sqrt((x[-1] - x[0])**2 + (y[-1] - y[0])**2)
        return total_distance / euclidean_dist if euclidean_dist > 0 else np.nan
    return np.nan

df['mouse_total_distance'] = df.apply(lambda row: calc_total_distance(row['x'], row['y']), axis=1)
df['mouse_path_linearity'] = df.apply(lambda row: calc_path_linearity(row['x'], row['y']), axis=1)

```

```python
import pandas as pd
import numpy as np

# Step 1: Calculate mouse_bounding_box_area (if not already done)
def calc_bounding_box_area(x, y):
    if isinstance(x, list) and isinstance(y, list) and len(x) >= 2:
        return (max(x) - min(x)) * (max(y) - min(y))
    return np.nan

df['mouse_bounding_box_area'] = df.apply(
    lambda row: calc_bounding_box_area(row['x'], row['y']), axis=1
)

# Step 2: Define function to extract screen bounds
def get_screen_bounds(group):
    all_x = sum(group['x'].dropna().tolist(), [])
    all_y = sum(group['y'].dropna().tolist(), [])
    return pd.Series({
        'screen_min_x': min(all_x) if all_x else np.nan,
        'screen_max_x': max(all_x) if all_x else np.nan,
        'screen_min_y': min(all_y) if all_y else np.nan,
        'screen_max_y': max(all_y) if all_y else np.nan,
    })

# Step 3: Drop existing screen bounds columns if any
df = df.drop(columns=[
    'screen_min_x', 'screen_max_x',
    'screen_min_y', 'screen_max_y'
], errors='ignore')

# Step 4: Group only by x and y (not profile_guid itself)
bounds_df = (
    df.groupby('profile_guid')[['x', 'y']]
    .apply(get_screen_bounds)
    .reset_index()
)

# Step 5: Merge screen bounds into the main DataFrame
df = df.merge(bounds_df, on='profile_guid', how='left')

# Step 6: Normalize bounding box area
def normalize_area(row):
    if pd.isna(row['mouse_bounding_box_area']):
        return np.nan
    width = row['screen_max_x'] - row['screen_min_x']
    height = row['screen_max_y'] - row['screen_min_y']
    if width > 0 and height > 0:
        screen_area = width * height
        return row['mouse_bounding_box_area'] / screen_area
    return np.nan

df['mouse_bounding_box_normalized'] = df.apply(normalize_area, axis=1)
```

```python
def calc_direction_changes(array):
    if not isinstance(array, list) or len(array) < 3:
        return np.nan
    diffs = np.diff(array)
    signs = np.sign(diffs)
    changes = np.diff(signs)
    return np.sum(changes != 0)

df['direction_changes_x'] = df['x'].apply(calc_direction_changes)
df['direction_changes_y'] = df['y'].apply(calc_direction_changes)
```

```python
import numpy as np
from collections import Counter
from math import log2

# Helper: Calculate direction entropy
def calc_direction_entropy(x, y):
    if not isinstance(x, list) or not isinstance(y, list) or len(x) < 2:
        return np.nan

    dx = np.diff(x)
    dy = np.diff(y)

    directions = []
    for dxi, dyi in zip(dx, dy):
        if dxi == 0 and dyi == 0:
            continue
        angle = np.arctan2(dyi, dxi)
        # Convert angle to 8-bin direction label
        bucket = int((angle + np.pi) / (2 * np.pi) * 8) % 8
        directions.append(bucket)

    if not directions:
        return 0.0

    count = Counter(directions)
    total = len(directions)
    entropy = -sum((c / total) * log2(c / total) for c in count.values())
    return entropy

# Helper: Calculate idle ratio
def calc_idle_ratio(x, y):
    if not isinstance(x, list) or not isinstance(y, list) or len(x) < 2:
        return np.nan
    dx = np.diff(x)
    dy = np.diff(y)
    idle_count = np.sum((dx == 0) & (dy == 0))
    return idle_count / len(dx)

# Helper: Calculate average speed
def calc_avg_speed(x, y, interval_sec=0.2):  # 200 ms = 0.2s per step
    if not isinstance(x, list) or not isinstance(y, list) or len(x) < 2:
        return np.nan
    dx = np.diff(x)
    dy = np.diff(y)
    distances = np.sqrt(dx**2 + dy**2)
    moving = distances > 0
    if not np.any(moving):
        return 0.0
    return np.mean(distances[moving]) / interval_sec

# Helper: Calculate average acceleration
def calc_avg_acceleration(x, y, interval_sec=0.2):
    if not isinstance(x, list) or not isinstance(y, list) or len(x) < 3:
        return np.nan
    dx = np.diff(x)
    dy = np.diff(y)
    distances = np.sqrt(dx**2 + dy**2)
    speeds = distances / interval_sec
    acc = np.diff(speeds) / interval_sec
    return np.mean(np.abs(acc))

# Apply to DataFrame
df['mouse_idle_ratio'] = df.apply(lambda row: calc_idle_ratio(row['x'], row['y']), axis=1)
df['mouse_average_speed'] = df.apply(lambda row: calc_avg_speed(row['x'], row['y']), axis=1)
df['mouse_average_acceleration'] = df.apply(lambda row: calc_avg_acceleration(row['x'], row['y']), axis=1)
df['mouse_direction_entropy'] = df.apply(lambda row: calc_direction_entropy(row['x'], row['y']), axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric feature columns
feature_cols = [
    'mouse_bounding_box_area',
    'mouse_bounding_box_normalized',
    'mouse_idle_ratio',
    'mouse_average_speed',
    'mouse_average_acceleration',
    'mouse_direction_entropy'
]

# Drop rows with NaN in any of the features
df_corr = df[feature_cols].dropna()

# Compute correlation matrix
corr_matrix = df_corr.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title("Correlation Matrix of Mouse Features")
plt.tight_layout()
plt.show()
```

```python
from sklearn.feature_selection import VarianceThreshold

# Features only
X = df[[
    'mouse_bounding_box_area',
    'mouse_bounding_box_normalized',
    'mouse_idle_ratio',
    'mouse_average_speed',
    'mouse_average_acceleration',
    'mouse_direction_entropy'
]].dropna()

# Drop features with zero or near-zero variance
selector = VarianceThreshold(threshold=0.001)
X_reduced = selector.fit_transform(X)
selected_features = X.columns[selector.get_support()]
print("Features with sufficient variance:", list(selected_features))
```

```python
# Compute correlation matrix again
corr = X[selected_features].corr().abs()

# Select upper triangle
upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

# Drop one of each pair of features with correlation > 0.85
to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]

print(" Highly correlated (to drop):", to_drop)

final_features = [f for f in selected_features if f not in to_drop]
print(" Final selected features:", final_features)
```
