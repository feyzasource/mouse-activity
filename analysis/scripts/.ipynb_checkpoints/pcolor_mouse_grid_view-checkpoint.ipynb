```

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

# CSV dosyasını yükle (dosya yolunu kendin ayarla)
df = pd.read_csv("real_activity_log.csv")

# Seçilen profile_guid
profile_guid = "ff7e90b7-06bc-461f-9b28-15687c010748"
df = df[df["profile_guid"] == profile_guid].copy()

# Liste string'lerini diziye çevir
def parse_array(text):
    try:
        return list(map(int, text.strip("{}").split(",")))
    except:
        return []

df["x_list"] = df["x"].apply(parse_array)
df["y_list"] = df["y"].apply(parse_array)

df = df[df["x_list"].apply(len) > 0]
df = df[df["y_list"].apply(len) > 0]

df["start_date_time"] = pd.to_datetime(df["start_date_time"], errors="coerce")
df = df[df["start_date_time"].notna()]
df["hour"] = df["start_date_time"].dt.floor("H")

# Grid boyutu (20x20 px hücreler)
grid_width = 96
grid_height = 54
scale_x = 20
scale_y = 20

# Örnek bir saat seç
selected_hour = df["hour"].iloc[0]
df_hour = df[df["hour"] == selected_hour]

# Heatmap başlat
heatmap = np.zeros((grid_height, grid_width), dtype=int)

# Heatmap doldur
for _, row in df_hour.iterrows():
    x_vals = row["x_list"]
    y_vals = row["y_list"]
    for x, y in zip(x_vals, y_vals):
        x_reduced = x // scale_x
        y_reduced = y // scale_y
        if 0 <= x_reduced < grid_width and 0 <= y_reduced < grid_height:
            heatmap[y_reduced, x_reduced] += 1

# PColor ile kutucuk görseli
plt.figure(figsize=(12, 6))
plt.pcolor(heatmap, cmap='Greys', edgecolors='k', linewidths=0.2)
plt.gca().invert_yaxis()
plt.xlabel("X Grid")
plt.ylabel("Y Grid")
plt.title(f"PColor Kutucuk Haritası - {selected_hour}")
plt.tight_layout()
plt.show()
```
