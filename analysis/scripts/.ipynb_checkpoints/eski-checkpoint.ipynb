```python
import pandas as pd
import ast
import numpy as np

# CSV y√ºkle
df = pd.read_csv("C:\\store\\git\\km-stat-activity\\data\\real\\real_activity_log.csv")  # <- dosya adƒ±nƒ± gerektiƒüi gibi deƒüi≈ütir

# x ve y'yi g√ºvenle listeye √ßevir
def safe_to_list(val):
    try:
        return list(ast.literal_eval(val)) if pd.notna(val) else []
    except (ValueError, SyntaxError):
        return []

df['x'] = df['x'].apply(safe_to_list)
df['y'] = df['y'].apply(safe_to_list)

# Ardƒ±ndan tekrar string olarak [] formatƒ±nda kaydet (kalƒ±cƒ± olsun diye)
df['x'] = df['x'].apply(lambda l: str(l))
df['y'] = df['y'].apply(lambda l: str(l))

# Yeni dosyaya yaz
df.to_csv("C:\\store\\git\\km-stat-activity\\data\\real\\mouse_data_cleaned.csv", index=False)

print("Yeni dosya mouse_data_cleaned.csv olarak kaydedildi.")
```

```python
import pandas as pd
import ast

# ‚úÖ Yeni formatlƒ± CSV dosyasƒ±nƒ± y√ºkle (x, y verileri k√∂≈üeli parantez i√ßinde)
df = pd.read_csv("C:\\store\\git\\km-stat-activity\\data\\real\\mouse_data_cleaned.csv")  # ‚Üê senin dosyanƒ±n adƒ± bu olmalƒ±

# üîÑ x ve y array‚Äôlerini g√ºvenli bir ≈üekilde listeye √ßevir (zaten k√∂≈üeli parantezde)
def safe_to_list(val):
    try:
        return list(ast.literal_eval(val)) if pd.notna(val) else []
    except (ValueError, SyntaxError):
        return []

df['x'] = df['x'].apply(safe_to_list)
df['y'] = df['y'].apply(safe_to_list)

# üîÅ Y√∂n deƒüi≈üimlerini sayan fonksiyon
def count_direction_changes(arr):
    count = 0
    for i in range(2, len(arr)):
        diff1 = arr[i-1] - arr[i-2]
        diff2 = arr[i] - arr[i-1]
        if diff1 * diff2 < 0:
            count += 1
    return count

# üßÆ Bounding box alanƒ± hesaplayan fonksiyon
def bounding_box_area(x_arr, y_arr):
    if not x_arr or not y_arr:
        return 0
    width = max(x_arr) - min(x_arr)
    height = max(y_arr) - min(y_arr)
    return width * height

# üìä Yeni √∂zellikleri hesapla
df['x_direction_changes'] = df['x'].apply(count_direction_changes)
df['y_direction_changes'] = df['y'].apply(count_direction_changes)
df['bounding_box_area'] = df.apply(lambda row: bounding_box_area(row['x'], row['y']), axis=1)

# üíæ Sonucu yeni bir CSV dosyasƒ±na kaydet
df.to_csv("C:\\store\\git\\km-stat-activity\\data\\real\\mouse_data_direction_bounding_box.csv", index=False)

print("Yeni veriseti ba≈üarƒ±yla kaydedildi: mouse_data_direction_bounding_box.csv")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# üîΩ Yerel dosya yolunu buraya girin
csv_path = r"C:\store\git\km-stat-activity\data\real\mouse_data_direction_bounding_box.csv"

# CSV dosyasƒ±nƒ± y√ºkle
df = pd.read_csv(csv_path)

# bounding_box_area sƒ±fƒ±r olanlarƒ± filtrele
df = df[df['bounding_box_area'] > 0]

# Kullanƒ±cƒ± ba≈üƒ±na ortalama bounding_box_area hesapla
user_means = df.groupby('profile_guid')['bounding_box_area'].mean().reset_index()
user_means.rename(columns={'bounding_box_area': 'avg_bounding_box_area'}, inplace=True)

# Normalize et
scaler = StandardScaler()
scaled_data = scaler.fit_transform(user_means[['avg_bounding_box_area']])

# KMeans k√ºmeleme (3 k√ºme)
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
user_means['cluster'] = kmeans.fit_predict(scaled_data)

# Heatmap verisini hazƒ±rla
heatmap_data = user_means.sort_values(by='avg_bounding_box_area')
heatmap_matrix = np.expand_dims(heatmap_data['avg_bounding_box_area'].values, axis=1)

# Heatmap √ßiz
plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_matrix, cmap='YlOrRd', yticklabels=False, cbar_kws={'label': 'Average Bounding Box Area'})
plt.title("Users Clustered by Average Bounding Box Area")
plt.xlabel("Bounding Box Area")
plt.ylabel("Users (sorted)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# CSV dosyasƒ±nƒ± y√ºkle (dosya yolunu kendine g√∂re ayarla)
csv_path = r"C:\store\git\km-stat-activity\data\real\mouse_data_direction_bounding_box.csv"
df = pd.read_csv(csv_path)

# bounding_box_area sƒ±fƒ±r olanlarƒ± √ßƒ±kar
df = df[df['bounding_box_area'] > 0]

# Kullanƒ±cƒ±larƒ±n dakikalƒ±k bounding_box_area verisini pivotla
pivot_df = df.pivot(index='start_date_time', columns='profile_guid', values='bounding_box_area')

# Eksik deƒüerleri 0 ile doldur
pivot_df = pivot_df.fillna(0)

# Transpose ederek satƒ±r = kullanƒ±cƒ±, s√ºtun = dakika yap
user_df = pivot_df.T

# Normalize et
scaler = StandardScaler()
X_scaled = scaler.fit_transform(user_df)

# K-means ile k√ºmelere ayƒ±r (k=4 √∂rnek olarak se√ßildi, deƒüi≈ütirilebilir)
kmeans = KMeans(n_clusters=4, random_state=42)
user_df['cluster'] = kmeans.fit_predict(X_scaled)

# K√ºmelenmi≈ü kullanƒ±cƒ±larƒ± CSV olarak dƒ±≈üa aktar
user_df.to_csv("clustered_users_by_boundingbox.csv")

# K√ºme sonu√ßlarƒ±nƒ± √∂zetle
print(user_df['cluster'].value_counts())

# K√ºme ortalamalarƒ±nƒ± g√∂rselle≈ütir (isteƒüe baƒülƒ±)
user_df.drop('cluster', axis=1).groupby(user_df['cluster']).mean().T.plot(figsize=(14, 5), legend=True)
plt.title("Bounding Box Area Averages per Cluster")
plt.xlabel("Time (Minutes)")
plt.ylabel("Normalized Bounding Box Area")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns

# CSV dosyasƒ±nƒ± y√ºkle
csv_path = r"C:\store\git\km-stat-activity\data\real\mouse_data_direction_bounding_box.csv"
df = pd.read_csv(csv_path)

# bounding_box_area sƒ±fƒ±r olanlarƒ± √ßƒ±kar
df = df[df['bounding_box_area'] > 0]

# Kullanƒ±cƒ±larƒ±n dakikalƒ±k bounding_box_area verisini pivotla
pivot_df = df.pivot(index='start_date_time', columns='profile_guid', values='bounding_box_area')

# Eksik deƒüerleri 0 ile doldur
pivot_df = pivot_df.fillna(0)

# Transpose ederek kullanƒ±cƒ±larƒ± satƒ±ra al
user_df = pivot_df.T

# Normalize et
scaler = StandardScaler()
X_scaled = scaler.fit_transform(user_df)

# KMeans k√ºmeleme (√∂rnek: 10 k√ºme)
k = 10
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# t-SNE ile boyut indirgeme (2D)
tsne = TSNE(n_components=2, perplexity=30, n_iter=3000, random_state=42)
X_2d = tsne.fit_transform(X_scaled)

# G√∂rselle≈ütir
plt.figure(figsize=(10, 7))
palette = sns.color_palette("tab10", k)
sns.scatterplot(x=X_2d[:, 0], y=X_2d[:, 1], hue=clusters, palette=palette, legend="full", s=50)
plt.title("User Clusters Based on Bounding Box Area (t-SNE + KMeans)")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```
