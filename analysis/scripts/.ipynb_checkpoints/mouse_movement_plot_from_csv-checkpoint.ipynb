# Mouse Movement Path from CSV
Bu notebook, bir kullanıcının seçilen dakikalık mouse hareketlerini CSV dosyasından okuyarak X-Y düzleminde çizer.

```python
# 📦 Gerekli kütüphaneler
import pandas as pd
import matplotlib.pyplot as plt
import ast
```

```python
# 📄 CSV dosyasını yükle
csv_path = "../data/activity_data.csv"  # Dosya yolu klasör yapınıza göre değiştirin
df = pd.read_csv(csv_path)
```

```python
# 🔍 Kullanıcı ve zaman filtrelemesi
profile_guid = "002e8f20-d5c2-4153-b1d0-9e1332549ac6"
start_time = "2025-04-14 08:10:47"

filtered = df[(df['profile_guid'] == profile_guid) & (df['start_date_time'] == start_time)]

if filtered.empty:
    print("Seçilen kullanıcı ve zaman için veri bulunamadı.")
else:
    row = filtered.iloc[0]

    # 🔢 Koordinatları güvenli şekilde parse et
    x_coords = ast.literal_eval(row['x'])
    y_coords = ast.literal_eval(row['y'])

    # 🎯 Grafik çizimi
    plt.figure(figsize=(10, 6))
    plt.plot(x_coords, y_coords, marker='o', linestyle='-')
    plt.title(f"Mouse Movement Path\n{profile_guid} @ {start_time}")
    plt.xlabel("X Coordinate")
    plt.ylabel("Y Coordinate")
    plt.gca().invert_yaxis()
    plt.grid(True)
    plt.show()
```
